```python
def stats(sheet, col):
    df = pd.read_csv (sheet)
    mean = df[col].mean()
    max_ = df[col].max()
    min_ = df[col].min()
    median = df[col].median() 
    std = df[col].std() 
    var = df[col].var()  
    
    print(col)
    print ('Mean '  , (mean))
    print ('Max '  , (max_))
    print ('Min '  , (min_))
    print ('Median ' , (median))
    print ('Std ' , (std))
    print ('Var ' , (var))
```

```python
for sheet in sheet_list:
    print("\n\n" + sheet)
    for col in cols:
        stats(sheet, col)
```
